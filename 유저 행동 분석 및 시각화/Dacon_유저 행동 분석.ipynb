### 특정 월 급증 이유 탐색 - 대회 정보 확인

```python
# 대회 정보 확인

# '0000-00'
def year_month(x):
    x = str(x)
    return x[:7]
# year_month
info_compet['period_start_year_month'] = info_compet['period_start'].apply(year_month)
info_compet['period_end_year_month'] = info_compet['period_end'].apply(year_month)
info_compet['merge_deadline_year_month'] = info_compet['merge_deadline'].apply(year_month)
```

```python
# 대회 시작/종료일, 팀 병합 마감일 year_month별 count
tmp1 = info_compet.groupby(['period_start_year_month'])[['cpt_id']].count().reset_index()
tmp2 = info_compet.groupby(['period_end_year_month'])[['cpt_id']].count().reset_index()
tmp3 = info_compet.groupby(['period_end_year_month'])[['cpt_id']].count().reset_index()
tmp1.columns = ['year_month','period_start']
tmp2.columns = ['year_month','period_end']
tmp3.columns = ['year_month','merge_deadline']
```

```python
tmp1['year_month'] = tmp1['year_month'].astype(str)
tmp2['year_month'] = tmp2['year_month'].astype(str)
tmp3['year_month'] = tmp3['year_month'].astype(str)
```

```python
compet_cnt_year_month = pd.merge(tmp1, tmp2, on='year_month', how='outer')
compet_cnt_year_month = pd.merge(compet_cnt_year_month, tmp3, on='year_month', how='outer').sort_values(by='year_month')
compet_cnt_year_month.reset_index(inplace=True)
compet_cnt_year_month.drop('index', axis=1, inplace=True)
```

```python
compet_cnt_year_month.style.background_gradient(cmap='Blues')     # .style.bar()
```

## Monthly New / Return User
- 참고! https://brunch.co.kr/@userhabit/19
- 신규 방문자 = 해당 월에 첫 로그인한 사용자 (그 월에 재방문 했어도 신규 가입자로 취급)
- 재방문자 = 해당 월까지의 누적 로그인 2회 이상 (첫 로그인 이후 월 login부터)
- unique 사용자 수로 계산

```python
# year, year_month, week_num(몇주차인지), week, hour 추가
info_login['c_time'] = pd.to_datetime(info_login['c_time'])
info_login['year'] = info_login['c_time'].dt.year
info_login['month'] = info_login['c_time'].apply(lambda x : x.strftime('%Y-%m'))
info_login['week_num'] = info_login['c_time'].apply(lambda x : x.strftime('%W'))
info_login['week'] = info_login['c_time'].apply(lambda x : x.weekday())
info_login['hour'] = info_login['c_time'].dt.hour
info_login
```

```python
# unique user_id로 counting할 거기 때문에, user_id, month만 남기고 중복제거하면 됨
new_return = info_login[['user_id', 'month']].copy()
new_return.drop_duplicates(inplace=True)
new_return.columns = ['user_id', 'login_month']
new_return = new_return.sort_values(by=['user_id', 'login_month'])
new_return.reset_index(inplace=True)
new_return.drop('index', axis=1, inplace=True)
new_return
```

```python
# 월별 신규/재방문 여부 표시
new_return['new'] = 0
new_return['return'] = 0

for i in tqdm(range(1,len(new_return))):     # index 0은 따로 처리
    if new_return.loc[i, 'user_id'] != new_return.loc[i-1, 'user_id']:
        new_return.loc[i, 'new'] = 1
    else:
        new_return.loc[i, 'return'] = 1
```

```python
new_return.loc[0, 'new'] = 1
```

```python
new_return
```

```python
month_new_return = new_return.groupby(['login_month'])[['new','return']].sum().reset_index()
month_new_return
```

```python
month_new_return['total'] = month_new_return['new'] + month_new_return['return']
month_new_return['new_rate'] = month_new_return['new'] / (month_new_return['total']) * 100
month_new_return['return_rate'] = month_new_return['return'] / (month_new_return['total']) * 100

month_new_return['new_rate'] = month_new_return['new_rate'].apply(lambda x : round(x, 1)) # 반올림
month_new_return['return_rate'] = month_new_return['return_rate'].apply(lambda x : round(x, 1)) # 반올림
month_new_return
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(go.Bar(x=month_new_return['login_month'], y=month_new_return['new_rate'], text=month_new_return['new_rate'],textposition='auto', name = 'Monthly New User % (Unique)', marker_color='#c8c8c8'), secondary_y=False)
fig.add_trace(go.Bar(x=month_new_return['login_month'], y=month_new_return['return_rate'],text=month_new_return['return_rate'],textposition='auto', name='Monthly Returning User % (Unique)', marker_color='#464646'), secondary_y=False)
fig.add_trace(go.Scatter(x=month_new_return['login_month'], y=month_new_return['total'],name='Monthly User Count (Unique)'), secondary_y=True)

fig.update_layout(
                template='plotly_white', 
                xaxis = {'title':'year_month'},
                yaxis = {'title':'User Count'},
                title_text='Monthly New / Returning User',
                barmode='stack'
                 )
fig.show()
```

- 월별 순 방문자 수가 점차 증가하고 있음
- 그중 재방문자 비중이 점차 증가하는 경향을 보이며, 순 사용자 급증 월에는 신규 방문자 비중이 높아지는 경우가 많음
